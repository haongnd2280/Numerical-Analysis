```python
def rowExchange(Ab, n, k, i): 
    # Y tuong rat don gian: di tung cot tu 0 den n, thuc hien trao doi hai phan tu tren hai hang thuoc cot do. 
    for j in range(n + 1): 
        temp = Ab[k][j]
        Ab[k][j] = Ab[i][j]
        Ab[i][j] = temp  
    
    # Note: List in Python is mutator type, so we dont need to return list, it automatically save what we changed. 
```

```python
# Bat dau qua trinh the nguoc: 
def backSubs(Ab, n, ind, rankA): 
    x = [0 for i in range(n)]
    row = rankA - 1

    for i in range(n): 
        if i not in ind: # neu cot thu i khong thuoc ind. 
            print('Bien thu %d la bien tu do. ' %(i + 1))
            x[i] = float(input('Ban muon dat bien tu do thu %d bang: ' %(i + 1)))

    # Bat dau qua trinh khu. 
    for i in range(n - 1, -1, -1): 
        if i in ind:   # neu bien thu i + 1 la bien khu (pivot variable).
            dot = 0
            for j in range(i + 1, n): 
                dot += Ab[row][j] * x[j]
            
            x[i] = (Ab[row][n] - dot) / Ab[row][i]
            row -= 1
            
    return x  
```

```python
def printEliMatrix(Ab, m, n, ind, i, j):
    # In ra ma tran tam giac. 
    print('Elimination augmented matrix: ')
    for a in range(m): 
        for b in range(n + 1): 
            print('%.5f  ' %(Ab[a][b]), end='')
        print()

    print('index: ')
    print(ind)
    
    print('i = %d' %(i))
    print('j = %d' %(j))
    print()
```

```python
def Sol(Ab, ind, m, n, rankA, rankAb): 
    if rankA == m and rankA == n:     # square and invertible. (square full rank)
        print('He co nghiem duy nhat.')
        x = backSubs(Ab, n, ind, rankA)
    elif rankA == m and rankA < n:    # short and wide (full row rank)
        print('He co vo so nghiem.')
        x = backSubs(Ab, n, ind, rankA)
    elif rankA < m and rankA == n:    # tall and thin. (full column rank) 
        if rankA == rankAb: 
            print('He co nghiem duy nhat.')
            x = backSubs(Ab, n, ind, rankA)
        else: 
            return 'He vo nghiem.'
    elif rankA < m and rankA < n:     # not full rank. 
        if rankA == rankAb: 
            print('He co vo so nghiem.')
            x = backSubs(Ab, n, ind, rankA)
        else: 
            return 'He vo nghiem.'
            
    return x 
```

```python
def rankDeter(Ab, ind, m, n): 
    rankA = 0 
    for i in range(m):
        if ind[i] >= 0: 
            rankA += 1
    
    rankAb = rankA
    for i in range(m - 1, rankA - 1, -1): 
        if Ab[i][n] != 0: 
            rankAb = i + 1
            break 

    print('rankA = %d' %(rankA))
    print('rankAb = %d' %(rankAb))
    print()
    
    return rankA, rankAb
```

```python
def Gauss(Ab): 
    m = len(Ab)
    n = len(Ab[0]) - 1
    ind = [-1 for i in range(m)]   # ind[i] = j: phan tu khu thu i (hang i) o cot thu j.
    i, j = 0, 0 

    while i <= m - 1 and j <= n - 1:  # dung khi tim duoc het phan tu tru hoac di het cot.
        if Ab[i][j] != 0: 
            ind[i] = j   # luu phan tru tru thu i o cot thu j. 
            print('Khong can hoan doi.')
        else: 
            print('Can hoan doi.')
            # Tim hang de hoan doi. 
            for k in range(i + 1, m): 
                if Ab[k][j] != 0:  
                    rowExchange(Ab, n, k, i)  # Hoan doi hang thu k va hang thu i cho nhau.
                    ind[i] = j

                    print('Row %d and row %d are exchanged!' %(k + 1, i + 1))
                    break  
        
        # Kiem tra xem ind[i] = j hay khong, neu co thi bat dau khu cot thu j, neu khong thi sang cot tiep theo. 
        if ind[i] == j:  # (i, j) chinh la phan tu tru (pivot element) thu i (hang i, cot j).
            for k in range(i + 1, m): 
                Ab[k][j] = Ab[k][j] / Ab[i][j]   
                
                for t in range(j + 1, n + 1): 
                    Ab[k][t] = Ab[k][t] - Ab[k][j] * Ab[i][t]

            # dat lai cac he so bang 0 o cot thu j (cho de nhin).
            for t in range(i + 1, m): 
                Ab[t][j] = 0  

            i += 1 
            j += 1  

            print('Ma tran Ab sau khi khu phan tu tru thu %d o cot thu %d: ' %(i, j))
            for a in range(m): 
                for b in range(n + 1): 
                    print('%.5f     ' %(Ab[a][b]), end='')
                print()
            print()
        else: 
            # neu khong tim duoc phan tu khu o cot thu j thi tang j them mot don vi, phan tu tru thu i giu nguyen. 
            print('Khong hoan doi duoc hang nao tren cot thu %d' %(j + 1))
            print()
            j += 1   
          
    printEliMatrix(Ab, m, n, ind, i, j)
    
    rankA, rankAb = rankDeter(Ab, ind, m, n)

    x = Sol(Ab, ind, m, n, rankA, rankAb)
    
    if x == 'He vo nghiem.': 
        print(x)
    else: 
        print('Nghiem cua he Ax = b la: ')
        print(x)
```

```python
Ab1 = [[1, 1, 0, 3, 4], 
      [2, 1, -1, 1, 1], 
      [3, -1, -1, 2, -3],
      [-1, 2, 3, -1, 4]]

if __name__ == '__main__':
    Gauss(Ab1)               # [-1, 2, 0, 1]
```
