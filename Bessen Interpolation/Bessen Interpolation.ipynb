```python
 import math 

def HornerInverseMethod(X):
    if X == []: 
        W = [1]
    else: 
        W = []
        W.append(1)
        W.append(-X[0])
    
        for i in range(1, len(X)):             
            ithMulti = [X[i] * k for k in W]
        
            W.append(0)   
            for i in range(1, len(W)): 
                W[i] = W[i] - ithMulti[i - 1]
            
    return W

def polyEval(x, Pn): 
    Q = []            
    Q.append(Pn[0])    
    for i in range(1, len(Pn)): 
        y = Pn[i] + Q[i - 1] * x 
        Q.append(y)
        
    return y

def factorial(n): 
    if n == 0: 
        return 1 
    return n*factorial(n - 1)

def difference(X, Y): 
    n = len(X)
    
    difTab = []
    # them cac mang rong vao hang. 
    for i in range(n): # i chay tu 0 den n - 1. 
        row_i = []
        for j in range(i + 2): 
            row_i.append(0)
        
        difTab.append(row_i)
    
    # Bat dau nhap cac vector cot cua X va Y. 
    for i in range(n): 
        difTab[i][0] = X[i]
    for i in range(n): 
        difTab[i][1] = Y[i]
        
    for e in difTab: 
        print(e)
    print()
    
    # Bat dau tinh sai phan cac cap. 
    # ta tinh lan luot theo tung cot (co dinh cot). 
    for j in range(2, n + 1): 
        for i in range(j - 1, n): 
            difTab[i][j] = difTab[i][j - 1] - difTab[i - 1][j - 1]
            
    for e in difTab: 
        print(e)
    print()
    return difTab 

def Coefs(n1): 
    T = []
    for i in range(0, n1 - 1):   
        T.append((0.5 + i)**2)
    print(T)
        
    L = []
    for i in range(len(T) + 1): 
        Li = HornerInverseMethod(T[0:i])
        L.append(Li)
    
    print("Cac mang he so thu duoc theo thuat toan Horner nguoc la (bien u^2): ")
    print()
    for e in L: 
        print(e)
    print()
    
    return L 

def main(X, Y, x): 
    h = X[1] - X[0]
    n = len(X)  # so diem moc.
    # Dem so moc noi suy nam o ben trai va ben phai cua x, va diem noi suy gan x nhat o ben trai cung nhu o ben phai.  
    lcount = math.floor((x - X[0])/h) + 1 
    rcount = (n - lcount)
    
    lpoint = X[0] + (lcount - 1)*h   # (lcount - 1) la so khoang truoc x. 
    rpoint = X[0] + lcount*h  
    
    print("So diem moc o ben trai diem x can tinh la: %s" % (lcount))
    print("Diem moc gan x nhat o ben trai la: %s" % (lpoint))
    print("So diem moc o ben phai diem x can tinh la: %s" % (rcount))
    print("Diem moc gan x nhat o ben phai la: %s" % (rpoint))
    
    # Chon so moc noi suy muon su dung.
    n = 1
    while n % 2 != 0: 
        print("Ban muon su dung bao nhieu moc noi suy (nhap vao mot so chan): ")
        n = int(input())
    
    X1 = []
    Y1 = []
    
    n1 = math.floor(n/2)
    k = (lcount - 1) - (n1 - 1)  # chi so cua so hạng đầu tiên. 
    for i in range(k, k + n): 
            X1.append(X[i])
            Y1.append(Y[i])
    
    # Lap bang sai phan cua bien t^2: t^2(t^2 - 1^2)(t^2 - 2^2)...
    difTab = difference(X1, Y1)
    
    # Cac mang he so Li, bien u^2.
    L2 = Coefs(n1) # L2 co do dai lẻ. 
    
    # Cho L2 va L1 vao cung mot mang. 
    L = []
    for i in range(len(X1)): 
        Li = []
        for j in range(i + 1): 
            Li.append(0)
        L.append(Li)

    for e in L: 
        print(e)
    print()

    # Thêm L2 vào mảng L.
    for i in range(0, len(L2)): 
        for j in range(0, len(L2[i])): 
            L[i*2][j*2] = L2[i][j]
            
    for e in L: 
        print(e)
    print()
    
    # Thêm L1 vào mảng L để hoàn thiện. (update xuống các phần tử của mảng trước. )
    for i in range(1, len(L), 2):
        for j in range(len(L[i - 1])): 
            L[i][j] = L[i - 1][j]
            
    for e in L: 
        print(e)
    print()

        
    # Cho cac delta(y) vao rieng mot mang rieng de thuan tinh toan. 
    diag = []
    i = n1 - 1
    
    for j in range(1, len(difTab) + 1): 
        if j % 2 == 1: 
            diag.append((difTab[i][j] + difTab[i + 1][j])/2)
            i += 1
        else: 
            diag.append(difTab[i][j])
    
    print("Cac gia tri sai phan duoc su dung la: ")
    print(diag)
    print()
    
     # Nhan bang sai phan difTab voi mang he so L. 
    for i in range(len(L)): 
        for j in range(len(L[i])): 
            L[i][j] = L[i][j]*diag[i]
            
    print("Update lai cac mang L sau khi da nhan voi duong cheo diag cua bang difTab: ")
    print()
    for e in L: 
        print(e)
    print()
    
    # Chia mang he so L cho giai thua sau khi da nhan tich vo huong. 
    for i in range(len(L)): 
        for j in range(len(L[i])): 
            L[i][j] = L[i][j]/factorial(i)
            
    print("Update lai cac mang L sau khi da nhan voi giai thua : ")
    print()
    for e in L: 
        print(e)
    print()
        
    # Tiep theo, cong cac he so tuong ung cua cac mang ma tran he so L. 
    inversePn = [] 
    for i in range(len(L)):
        xi = 0 
        for j in range(len(L) - i): 
            xi += L[i + j][j]
        
        inversePn.append(xi)
    
    # Dao lai cac he so su dung negative indexes. 
    Pn = []
    for i in range(-1, -len(inversePn) - 1, -1): 
        Pn.append(inversePn[i])
        
    # In Pn ra man hinh. 
    print("Sau khi da cong cac he so tuong ung (cong theo duong cheo chinh), ta co da thuc noi suy Sterling thu duoc la: ")
    print()
    print(Pn)
        
    # Tinh gia tri cua x. 
    t = (x - X1[n1 - 1])/h
    u = t - 0.5
    y = polyEval(u, Pn)
    print("Tai x = %s, t = %s => u = %s => y = %s " % (x, t, u, y))
    
    return Pn
```

```python
X = [1.30, 1.35, 1.40, 1.45]
Y = [0.89747, 0.89115, 0.88726, 0.88565]

print("Nhap vao diem x ban muon tinh gia tri: ")
x = float(input())

main(X, Y, x)
```
