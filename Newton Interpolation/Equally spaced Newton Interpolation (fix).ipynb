# DL ban đầu: Một mảng hệ số rất dài. 
    
Bước 1: Nhập vào điểm x muốn tính giá trị. 
Bước 2: Tính h, số điểm mốc n. 
Bước 3: Dem so moc noi suy nam o ben trai va ben phai cua x, va diem noi suy gan x nhat o ben trai cung nhu o ben phai. 
Bước 4: Nhap vao lua chon tinh da thuc noi suy Newton tien hay lui. 
Bước 5: Chon so moc noi suy muon su dung (dựa vào bước 4)
Bước 6: Thêm các mốc cần tính vào mảng X1, Y1 (2 TH)
Bước 7: Bảng sai phân (X1, Y1)
Bước 7: Mảng hệ số L(n, choice)
Bước 8: Nhan bang sai phan difTab voi bang he so L (2 TH) 
Bước 9: Chia mảng L cho giai thừa.
Bước 10: Tiep theo, cong cac he so tuong ung cua cac mang ma tran he so L. 
Bước 11: Dao lai cac he so su dung negative indexes.
Bước 12: Tinh gia tri cua x, ban dau chuyển x về t trước. (2 TH) 

```python
# Ta tinh chinh lai mot chut thuat toan Horner nguoc tu bai da thuc noi suy lagrange de no co the tinh duoc mang X = empty. 
def HornerInverseMethod(X):
    if X == []: 
        W = [1]
    else: 
        W = []
        W.append(1)
        W.append(-X[0])
    
        for i in range(1, len(X)):             
            ithMulti = [X[i] * k for k in W]
        
            W.append(0)   
            for i in range(1, len(W)): 
                W[i] = W[i] - ithMulti[i - 1]
            
    return W

def factorial(n): 
    if n == 0: 
        return 1 
    return n*factorial(n - 1)

# Ham thu 2: Cong thuc tinh ty sai phan. 
def difference(X, Y): 
    n = len(X)
    
    difTab = []
    for i in range(n): 
        difTab.append([])
        
    for i in range(n):  
        for j in range((i + 1) + 1): 
            difTab[i].append(0)
    
    for i in difTab: 
        print(i)
    print()
    
    for i in range(n): 
        difTab[i][0] = X[i]
    for i in range(n): 
        difTab[i][1] = Y[i]
        
    for i in difTab: 
        print(i)
    print()
    
    # Ta bat dau tinh ty sai phan tu day (ke tu cot thu 3). 
    
    # di theo tung cot mot, vi ta tinh theo tung cot truoc. chi so cot bat dau tu 2.
    for j in range(2, n + 1): 
        # tu cot (fixed), ta xac dinh chi so hang de tinh tung phan tu mot.
        for i in range(j - 1, n): 
            difTab[i][j] = difTab[i][j - 1] - difTab[i - 1][j - 1]
    
    for e in difTab: 
        print(e)
    print()
    return difTab

# Tiep theo, la mot mang L chua cac mang con la cac mang he so cua cac thuat toan Horner nguoc. 
def Coefs(X, choice): 
    T = []
    if choice == 1: 
        for i in range(len(X)): 
            T.append(i)
    if choice == 2: 
        for i in range(len(X)): 
            T.append(-i)
    print(T)
        
    L = []
    for i in range(len(T)): 
        Li = HornerInverseMethod(T[0:i])
        L.append(Li)
    
    print("Cac mang he so thu duoc theo thuat toan Horner nguoc la: ")
    print()
    for e in L: 
        print(e)
    print()
    
    return L 

# Ham thuc thi chinh cua chuong trinh. 
def main(X, Y, x): 
    h = X[1] - X[0]
    n = len(X)  # so diem moc.
    # Dem so moc noi suy nam o ben trai va ben phai cua x, va diem noi suy gan x nhat o ben trai cung nhu o ben phai.  
    lcount = math.floor((x - X[0])/h) + 1 
    rcount = (n - lcount)
    
    lpoint = X[0] + (lcount - 1)*h   # (lcount - 1) la so khoang truoc x. 
    rpoint = X[0] + lcount*h  
    
    print("So diem moc o ben trai diem x can tinh la: %s" % (lcount))
    print("Diem moc gan x nhat o ben trai la: %s" % (lpoint))
    print("So diem moc o ben phai diem x can tinh la: %s" % (rcount))
    print("Diem moc gan x nhat o ben phai la: %s" % (rpoint))
    
    # Nhap vao lua chon tinh da thuc noi suy Newton tien hay lui. 
    choice = 0 
    while choice != 1 and choice != 2:   
        print("Ban muon tinh da thuc noi suy Newton tien hay lui: ")
        print("Tien: Nhan 1. ")
        print("Lui: Nhan 2. ")
        
        choice = int(input())
    
    # Chon so moc noi suy muon su dung. 
    print("Ban muon su dung bao nhieu moc noi suy: ")
    n = int(input())
    
    # Thêm các mốc cần tính vào mảng X1, Y1. 
    X1 = []
    Y1 = []
    if choice == 1: 
        k = lcount - 1
        for i in range(k, k + n): 
            X1.append(X[i])
            Y1.append(Y[i])
    if choice == 2: 
        k = lcount - n + 1
        for i in range(k, k + n): 
            X1.append(X[i])
            Y1.append(Y[i])

    # bang ty sai phan.
    difTab = difference(X1, Y1)
    # cac mang he so Li. 
    L = Coefs(X1, choice)
    
    # Nhan bang sai phan difTab voi bang he so L. 
    # Newton tien: 
    if choice == 1: 
        for i in range(len(difTab)): 
            L[i] = [k * difTab[i][i + 1] for k in L[i]] 
    # Newton lùi. 
    if choice == 2: 
        # hang cuoi cung fixed => i = (len(difTab) - 1) 
        i = len(difTab) - 1  # co dinh i.
        for j in range(len(L)): # mang he so L thu j, j = (0, len(L) - 1).
            for k in range(len(L[j])): 
                L[j][k] = L[j][k] * difTab[i][j + 1]
        
    print("Update lai cac mang L sau khi da nhan voi bang difTab: ")
    print()
    for e in L: 
        print(e)
    print()
    
    # Chia mảng L cho giai thừa. 
    for i in range(len(L)): 
        for j in range(len(L[i])): 
            L[i][j] = L[i][j]/factorial(i)
            
    print("Update lai cac mang L sau khi da nhan voi giai thua : ")
    print()
    for e in L: 
        print(e)
    print()
    
    # Tiep theo, cong cac he so tuong ung cua cac mang ma tran he so L. 
    inversePn = [] 
    for i in range(len(L)):
        xi = 0 
        for j in range(len(L) - i): 
            xi += L[i + j][j]
        
        inversePn.append(xi)
    
    # Dao lai cac he so su dung negative indexes. 
    Pn = []
    for i in range(-1, -len(inversePn) - 1, -1): 
        Pn.append(inversePn[i])
    
    # In Pn ra man hinh. 
    print("Sau khi da cong cac he so tuong ung (cong theo duong cheo chinh), ta co da thuc noi suy Newton (tien/lui) thu duoc la: ")
    print()
    print(Pn)
    
    # Tinh gia tri cua x. 
    if choice == 1: 
        t = (x - X1[0])/h
    if choice == 2: 
        t = (x - X1[-1])/h 
        
    y = polyEval(t, Pn)
    print("Tai x = %s, t = %s => y = %s " % (x, t, y))
    
    return Pn

def polyEval(x, Pn): 
    Q = []            
    Q.append(Pn[0])    
    for i in range(1, len(Pn)): 
        y = Pn[i] + Q[i - 1] * x 
        Q.append(y)
        
    return y
```

```python
X = [1.35, 1.40, 1.45, 1.5, 1.55]
Y = [0.89115, 0.88726, 0.88565, 0.88623, 0.88887]
print("Nhap vao diem x ban muon tinh gia tri: ")
x = float(input())

Pn = main(X, Y, x)

for i in range(0, len(Pn)): 
    Pn[i] = Pn[i]*(len(Pn) - i - 1)
del Pn[-1]

for e in Pn: 
    print(e)
polyEval(0.6, Pn)
    
```
