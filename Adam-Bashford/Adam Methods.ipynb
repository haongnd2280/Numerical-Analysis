```python
# Adams Bashforth 
import numpy as np

def f(x, y): 
    return x + y

def factorial(n): 
    if n == 0: 
        return 1 
    return n*factorial(n - 1)

def difference(X, Y): 
    n = len(X)
    difTab = []
    
    # them cac mang rong vao hang. 
    for i in range(n): # i chay tu 0 den n - 1. 
        row_i = []
        for j in range(i + 2): 
            row_i.append(0)
        
        difTab.append(row_i)
    
    # Bat dau nhap cac vector cot cua X va Y. 
    for i in range(n): 
        difTab[i][0] = X[i]
    for i in range(n): 
        difTab[i][1] = Y[i]
    
    # Bat dau tinh sai phan cac cap. 
    # ta tinh lan luot theo tung cot (co dinh cot). 
    for j in range(2, n + 1): 
        for i in range(j - 1, n): 
            difTab[i][j] = difTab[i][j - 1] - difTab[i - 1][j - 1]
            
    return difTab 

def HornerInverseMethod(X):
    if X == []: 
        W = [1]
    else: 
        W = []
        W.append(1)
        W.append(-X[0])
    
        for i in range(1, len(X)):             
            ithMulti = [X[i] * k for k in W]
        
            W.append(0)   
            for i in range(1, len(W)): 
                W[i] = W[i] - ithMulti[i - 1]
            
    return W

# Newton lui
def Coefs(X): 
    T = []
    for i in range(len(X)): 
        T.append(-i)
        print(T)
        
    L = []
    for i in range(len(T)): 
        Li = HornerInverseMethod(T[0:i])
        L.append(Li)
    
    print("Cac mang he so thu duoc theo thuat toan Horner nguoc la: ")
    print()
    for e in L: 
        print(e)
    print()
    
    return L 

def RKthree(x0, y0, h, s, m): 
    p1 = float(input())
    p2 = float(input())
    p3 = float(input())
    alpha2 = float(input())
    alpha3 = float(input())
    beta21 = float(input())
    beta31 = float(input())
    beta32 = float(input())
    
    x = x0 
    y = y0
    
    X = np.array([])
    Y = np.array([])
    deriY = np.array([])
    
    X = np.append(X, x0)
    Y = np.append(Y, y0)
    deriY = np.append(deriY, f(x0, y0))
    
    for i in range(0, m - 1): 
        k1i = h*f(x, y)
        k2i = h*f(x + alpha2*h, y + beta21*k1i)
        k3i = h*f(x + alpha3*h, y + beta31*k1i + beta32*k2i)
        
        y = y + p1*k1i + p2*k2i + p3*k3i
        x = x0 + (i + 1)*h
        
        X = np.append(X, x)
        Y = np.append(Y, y)
        deriY = np.append(deriY, f(x, y))
        
        print(x)
        print(y)
        
        print()
        
    return X, Y, deriY

def RKfour(x0, y0, h, s, m): 
    p1 = 1/6
    p2 = 1/3
    p3 = 1/3
    p4 = 1/6
    alpha2 = 0.5
    alpha3 = 0.5
    alpha4 = 1
    beta21 = 0.5
    beta31 = 0
    beta32 = 0.5
    beta41 = 0
    beta42 = 0
    beta43 = 1
    
    x = x0 
    y = y0
    
    X = np.array([])
    Y = np.array([])
    deriY = np.array([])
    
    X = np.append(X, x0)
    Y = np.append(Y, y0)
    deriY = np.append(deriY, f(x0, y0))
    
    for i in range(0, m - 1): 
        k1i = h*f(x, y)
        k2i = h*f(x + alpha2*h, y + beta21*k1i)
        k3i = h*f(x + alpha3*h, y + beta31*k1i + beta32*k2i)
        k4i = h*f(x + alpha4*h, y + beta41*k1i + beta42*k2i + beta43*k3i)
        
        y = y + p1*k1i + p2*k2i + p3*k3i + p4*k4i 
        x = x0 + (i + 1)*h
        
        X = np.append(X, x)
        Y = np.append(Y, y)
        deriY = np.append(deriY, f(x, y))
        
        print(x)
        print(y)
        
        print()
        
    return X, Y, deriY

def AB(x0, y0, a, b, h, s, m): 
    if s == 3: 
        X, Y, deriY = RKthree(x0, y0, h, s, m)
    if s == 4: 
        X, Y, deriY = RKfour(x0, y0, h, s, m)
        
    # Tinh mang he so L. 
    L = Coefs(X)
    
    # tinh nguyen ham cua L, cung dong thoi la tinh tich phan can tu 0 den 1. 
    for i in range(len(L)): 
        n = len(L[i])
        for j in range(n): 
            L[i][j] = L[i][j]/(n - j)
        L[i].append(0)
        
    # Cong cac phan tu cua mang L[i] lai voi nhau, luu vao mot mang moi T. 
    T = np.array([])
    for i in range(len(L)): 
        Ti = 0 
        for j in range(len(L[i])): 
            Ti = Ti + L[i][j]
        T = np.append(T, Ti)
        
    # Chia mang he so T cho giai thua tuong ung. 
    for i in range(len(T)): 
        T[i] = T[i]/factorial(i)
    print(T)
    print()
    
    n = (b - a)/h 
    for i in range(0, int(n) - m + 1): 
        X1 = X[i:len(X)]
        deriY1 = deriY[i:len(deriY)]

        # Lap bang sai phan cua da deriY, lay m moc moi nhat. 
        difTab = difference(X1, deriY1)
        
        # Lay tich vo huong cua mang T va mang delta(y).  
        integral = 0
        for j in range(1, len(difTab[-1])): 
            integral += difTab[-1][j]*T[j - 1]
    
        # Ket qua cuoi cung. 
        y = Y[-1] + h*integral 
        x = x0 + (i + m)*h
        
        # Them x moi, y moi, y' moi vao mang X, Y, deriY. 
        X = np.append(X, x)
        Y = np.append(Y, y)
        deriY = np.append(deriY, f(x, y))
        
        print(x)
        print(y)
        print()
```

```python
# Adam Moulton. 

def Coefs(n): 
    T = []
    for i in range(n): 
        T.append(1 - i)
        print(T)
        
    L = []
    for i in range(len(T)): 
        Li = HornerInverseMethod(T[0:i])
        L.append(Li)
    
    print("Cac mang he so thu duoc theo thuat toan Horner nguoc la: ")
    print()
    for e in L: 
        print(e)
    print()
    
    return L 

def C(n, k): 
    return factorial(n)/(factorial(k)*factorial(n - k))
    
def AM(x0, y0, a, b, h, s, m): 
    if s == 3: 
        X, Y, deriY = RKthree(x0, y0, h, s, m - 1)
    if s == 4: 
        X, Y, deriY = RKfour(x0, y0, h, s, m - 1)
        
    # Tinh mang he so L. 
    L = Coefs(len(X) + 1)
    
    # tinh nguyen ham cua L, cung dong thoi la tinh tich phan can tu 0 den 1. 
    for i in range(len(L)): 
        n = len(L[i])
        for j in range(n): 
            L[i][j] = L[i][j]/(n - j)
        L[i].append(0)
        
    # Cong cac phan tu cua mang L[i] lai voi nhau, luu vao mot mang moi T. 
    T = np.array([])
    for i in range(len(L)): 
        Ti = 0 
        for j in range(len(L[i])): 
            Ti = Ti + L[i][j]
        T = np.append(T, Ti)
        
    # Chia mang he so T cho giai thua tuong ung. 
    for i in range(len(T)): 
        T[i] = T[i]/factorial(i)
    print(T)
    print()
    
    # Tinh mang he so B (nhi thuc)
    B = []
    for n in range(m): 
        Bi = []
        for k in range(n + 1): 
            Bi.append(C(n, k)*(-1)**k)
        B.append(Bi)
            
    for e in B: 
        print(e)
    print()
    
    # Nhan mang B voi mang T. 
    for i in range(len(B)): 
        for j in range(len(B[i])): 
            B[i][j] = B[i][j]*T[i]
            
    print("Mang B sau khi nhan voi mang T la: ")
    for e in B: 
        print(e)
    print()
    
    # Cong cac phan tu tuong ung cua mang B lai voi nhau, luu vao mang H. 
    H = []
    for j in range(len(B)): 
        Hi = 0 
        for i in range(j, len(B)): 
            Hi = Hi + B[i][j]
        H.append(Hi)
        
    print("Mang H cuoi cung thu duoc la: (he so cua cong thuc hien)")
    for e in H: 
        print(e)
    print()
        
     
    # Them x3. 
    X = np.append(X, x0 + (m - 1)*h)
    n = (b - a)/h  # so khoang cach deu nhau. 
    H1 = H[1:len(H)]
    
    print("Mang H1 la: ")
    print(H1)
    print()
    
    print("H[0] la: ")
    print(H[0])
    print()
    
    # Nhap vao sai so epsilon. 
    print("Nhap vao sai so epsilon mong muon: ")
    epsilon = float(input())
    print()
        
    # Bat dau vong lap.
    for i in range(0, int(n) - m + 2):  
        deriY1 = deriY[i:len(deriY)]
        
        print("Mang deriY1 la: ")
        print(deriY1)
        print()
        
        # Tinh sigma (lay dotproduct cua H1 va deriY1)
        sigma = 0 
        for i in range(len(deriY1)): 
            sigma += H1[i]*deriY1[len(deriY1) - i - 1]
        sigma = sigma*h 
        
        print("sigma la: ")
        print(sigma)
        print()
        
        # Tinh hai gia tri ban dau. 
        yt = Y[-1] + h*H[0]*f(X[-1], Y[-1]) + sigma
        ys = Y[-1] + h*H[0]*f(X[-1], yt) + sigma
        
        # Bat dau lap don. 
        while abs(ys - yt) >= epsilon: 
            yt = ys 
            ys = Y[-1] + h*H[0]*f(X[-1], yt) + sigma
            
        print(X[-1])
        print(ys)
        print()
        
        deriY = np.append(deriY, f(X[-1], ys))
        Y = np.append(Y, ys)
        X = np.append(X, X[-1] + h)
```
