```python
import math 

def HornerInverseMethod(X):
    if X == []: 
        W = [1]
    else: 
        W = []
        W.append(1)
        W.append(-X[0])
    
        for i in range(1, len(X)):             
            ithMulti = [X[i] * k for k in W]
        
            W.append(0)   
            for i in range(1, len(W)): 
                W[i] = W[i] - ithMulti[i - 1]
            
    return W

def polyEval(x, Pn): 
    Q = []            
    Q.append(Pn[0])    
    for i in range(1, len(Pn)): 
        y = Pn[i] + Q[i - 1] * x 
        Q.append(y)
        
    return y

def factorial(n): 
    if n == 0: 
        return 1 
    return n*factorial(n - 1)

# Ham thu 2: Cong thuc tinh ty sai phan. 
def difference(X, Y): 
    n = len(X)
    
    difTab = []
    for i in range(n): 
        difTab.append([])
        
    for i in range(n):  
        for j in range((i + 1) + 1): 
            difTab[i].append(0)
    
    for i in difTab: 
        print(i)
    print()
    
    for i in range(n): 
        difTab[i][0] = X[i]
    for i in range(n): 
        difTab[i][1] = Y[i]
        
    for i in difTab: 
        print(i)
    print()
    
    # Ta bat dau tinh ty sai phan tu day (ke tu cot thu 3). 
    
    # di theo tung cot mot, vi ta tinh theo tung cot truoc. chi so cot bat dau tu 2.
    for j in range(2, n + 1): 
        # tu cot (fixed), ta xac dinh chi so hang de tinh tung phan tu mot.
        for i in range(j - 1, n): 
            difTab[i][j] = difTab[i][j - 1] - difTab[i - 1][j - 1]
    
    for e in difTab: 
        print(e)
    print()
    return difTab

def Coefs(X, choice): 
    T = []
    if choice == 1: 
        for i in range(len(X)): 
            T.append(i)
    if choice == 2: 
        for i in range(len(X)): 
            T.append(-i)
    print(T)
        
    L = []
    for i in range(len(T)): 
        Li = HornerInverseMethod(T[0:i])
        L.append(Li)
    
    print("Cac mang he so thu duoc theo thuat toan Horner nguoc la: ")
    print()
    for e in L: 
        print(e)
    print()
    
    return L 

# Ham thuc thi chinh cua chuong trinh. 
def main(X, Y, y): 
    h = X[1] - X[0]
    n = len(X)  # so diem moc.
    
    # Chon diem noi suy x0. 
    print("Bạn muốn chọn mốc nội suy ban đầu x0 là điểm nào: (chú ý nhập vào thật chính xác điểm x0 có trong khoảng nội suy)")
    x0 = float(input())
    
    # Dem so moc noi suy nam o ben trai va ben phai cua x, va diem noi suy gan x nhat o ben trai cung nhu o ben phai.  
    lcount = math.floor((x0 - X[0])/h)
    rcount = (n - lcount - 1)
    
    lpoint = Y[0 + (lcount - 1)]   # (lcount - 1) la so khoang truoc x. 
    rpoint = Y[0 + lcount]
    
    print("Số điểm x ở bên trái điểm x0 bạn vừa nhập là: %s" % (lcount))
    print("Số điểm x ở bên phải điểm x0 bạn vừa nhập là: %s" % (rcount))
    
    # Nhap vao lua chon tinh da thuc noi suy Newton tien hay lui. 
    choice = 0 
    while choice != 1 and choice != 2:   
        print("Bạn muốn tính nội suy Newton ngược tiến hay lùi: ")
        print("Tiến: Nhấn 1. ")
        print("Lùi: Nhấn 2. ")
        
        choice = int(input())
    
    # Chon so moc noi suy muon su dung. 
    print("Bạn muốn sử dụng bao nhiêu mốc nội suy: ")
    n = int(input())
    
    X1 = []
    Y1 = []
    if choice == 1: 
        k = lcount
        for i in range(k, k + n): 
            X1.append(X[i])
            Y1.append(Y[i])
    if choice == 2: 
        k = lcount - n + 1
        for i in range(k, k + n): 
            X1.append(X[i])
            Y1.append(Y[i])

    # bang ty sai phan.
    difTab = difference(X1, Y1)
    # cac mang he so Li. 
    L = Coefs(X1, choice)
    
    # Update y0 = y - y0 hoac yn = y - yn. 
    if choice == 1: 
        difTab[0][1] = y - difTab[0][1]
    if choice == 2: 
        difTab[-1][1] = y - difTab[-1][1]
        
    # update L = -L. chi update tu hang thu 3 tro di. 
    for i in range(2, len(L)): 
        for j in range(len(L[i])): 
            L[i][j]= -L[i][j]
    # update L[1] = [0, 0]
    L[1][0] = 0
    
    print("Mảng L sau khi đã chuyển sang -L là: ")
    for e in L: 
        print(e)
    print()
    
    print("Update lại bảng difTab: ")
    for e in difTab: 
        print(e)
    print()

    # Nhan bang sai phan difTab voi bang he so L. 
    # Newton tien: 
    if choice == 1: 
        for i in range(len(difTab)): 
            L[i] = [k * difTab[i][i + 1] for k in L[i]] 
    if choice == 2: 
        # hang cuoi cung fixed => i = (len(difTab) - 1) 
        i = len(difTab) - 1  # co dinh i.
        for j in range(len(L)): # mang he so L thu j, j = (0, len(L) - 1).
            for k in range(len(L[j])): 
                L[j][k] = L[j][k] * difTab[i][j + 1]
        
    print("Update lại mảng L sau khi đã nhân với bảng difTab: ")
    print()
    for e in L: 
        print(e)
    print()
    
    # chia cho n!*delta(y0) hoặc n!*delta(y_(N-1))
    if choice == 1: 
        delta0 = difTab[1][2]
        for i in range(len(L)): 
            for j in range(len(L[i])): 
                L[i][j] = L[i][j]/(factorial(i)*delta0)
    if choice == 2: 
        deltaN = difTab[-1][2]
        for i in range(len(L)): 
            for j in range(len(L[i])): 
                L[i][j] = L[i][j]/(factorial(i)*deltaN)
            
    print("Update lại mảng L sau khi đã chia cho giai thừa và delta: ")
    print()
    for e in L: 
        print(e)
    print()
    
    # Tiep theo, cong cac he so tuong ung cua cac mang ma tran he so L. 
    inversePn = [] 
    for i in range(len(L)):
        xi = 0 
        for j in range(len(L) - i): 
            xi += L[i + j][j]
        
        inversePn.append(xi)
    
    # Dao lai cac he so su dung negative indexes. 
    Pn = []
    for i in range(-1, -len(inversePn) - 1, -1): 
        Pn.append(inversePn[i])
    
    # In Pn ra man hinh. 
    print("Sau khi đã cộng các hệ số tương ứng (cộng theo đường chéo chính), ta có đa thức nội suy Newton (tiến/lùi) thu được là: ")
    print()
    print(Pn)
    print()
    
    # Tim gia tri x tuong ung voi gia tri y da nhap vao. 
    # Xap xi dau. 
    if choice == 1: 
        t0 = (y - Y1[0])/difTab[1][2]
        print("t0 la: ", t0)
        print()
    if choice == 2: 
        t0 = (y - Y1[-1])/difTab[-1][2]
        print("t0 la: ", t0)
        print()
    
    # Bat dau vong lap, ta se su dung dieu kien epsilon, lay den 3 chu so sau dau phay. 
    print("Bạn muốn nhập epsilon là bao nhiêu: ")
    epsilon = float(input())
    print()
    
    t1 = polyEval(t0, Pn)
    t2 = polyEval(t1, Pn)
    loop = 2
    
    print("t0 là: ", t0)
    print("t1 là: ", t1)
    print("t2 là: ", t2)
    print()
    print("Bắt đầu vòng lặp: ")
    while abs(t2 - t1) >= epsilon: 
        t1 = t2 
        t2 = polyEval(t1, Pn)
        loop += 1 
        print("t = ", t2)
        
    print("Ta thu được giá trị t = %s sau %s lần tính (%s vòng lặp) " % (t2, loop, loop - 2))
    print()
    
    if choice == 1: # Newton tien
        x = X1[0] + h*t2
    if choice == 2: # Newton lui. 
        x = X1[-1] + h*t2 
        
    print("Giá trị x ứng với giá trị y bạn đã nhập vào ban đầu là: ", x )
```
