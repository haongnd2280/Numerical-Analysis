## Câu 2: Sử dụng Gauss Jordan

```python
Chú ý: 
- Ban đầu, chọn phần tử khử đầu tiên tại vị trí đang xét i, j.
- Tiếp, Chọn phần tử khử theo thứ tự 1, 2, 4, max, ...
- Chọn xong, kiểm tra xem phần tử khử có khác 0 hay không.
```

```python
def rowExchange(Ab, n, k, i): 
    # Y tuong: di tung cot tu 0 den n, thuc hien trao doi hai phan tu tren hai hang thuoc cot do. 
    for j in range(n + 1): 
        Ab[k][j], Ab[i][j] = Ab[i][j], Ab[k][j]
```

```python
def pivotSelection(Ab, m, i, j): 
    # Ưu tiên chọn phần tử khử là bội số của 2 theo thứ tự độ lớn tăng dần.
    for k in range(i, m): 
        if abs(Ab[k][j]) == 1: 
            return Ab[k][j], k
    for k in range(i, m): 
        if abs(Ab[k][j]) == 2: 
            return Ab[k][j], k
    for k in range(i, m): 
        if abs(Ab[k][j]) == 4: 
            return Ab[k][j], k
    for k in range(i, m): 
        if Ab[k][j] != 0 and Ab[k][j] % 2 == 0: 
            return Ab[k][j], k
    # Ưu tiên số hai là chọn phần tử khử là phần tử có trị tuyệt đối lớn nhất. 
    max = Ab[i][j]
    pivotRow = i
    for k in range(i + 1, m): 
        if abs(Ab[k][j]) > abs(max): 
            max = Ab[k][j]
            pivotRow = k
    
    return max, pivotRow
```

```python
def printEliMatrix(Ab, m, n, ind, i, j):
    # In ra ma tran tam giac. 
    print('Elimination augmented matrix: ')
    for a in range(m): 
        for b in range(n + 1): 
            print('%.5f  ' %(Ab[a][b]), end='')
        print()

    print('index: ')
    print(ind)
    
    print('i = %d' %(i))
    print('j = %d' %(j))
    print()
```

```python
def Sol(Ab, ind, m, n, rankA, rankAb): 
    if rankA == m and rankA == n:     # square and invertible. (square full rank)
        print('He co nghiem duy nhat.')
        x = solFinding(Ab, n, ind, rankA)
    elif rankA == m and rankA < n:    # short and wide (full row rank)
        print('He co vo so nghiem.')
        x = solFinding(Ab, n, ind, rankA)
    elif rankA < m and rankA == n:    # tall and thin. (full column rank) 
        if rankA == rankAb: 
            print('He co nghiem duy nhat.')
            x = solFinding(Ab, n, ind, rankA)
        else: 
            return 'He vo nghiem.'
    elif rankA < m and rankA < n:     # not full rank. 
        if rankA == rankAb: 
            print('He co vo so nghiem.')
            x = solFinding(Ab, n, ind, rankA)
        else: 
            return 'He vo nghiem.'
            
    return x 
```

```python
def rankDeter(Ab, ind, m, n): 
    rankA = 0 
    for i in range(m):
        if ind[i] >= 0: 
            rankA += 1
    
    rankAb = rankA
    for i in range(m - 1, rankA - 1, -1): 
        if Ab[i][n] != 0: 
            rankAb = i + 1
            break 

    print('rankA = %d' %(rankA))
    print('rankAb = %d' %(rankAb))
    print()
    
    return rankA, rankAb
```

```python
def solFinding(Ab, n, ind, rankA): 
    x = [0 for i in range(n)]
    row = rankA - 1

    for i in range(n): 
        if i not in ind: # neu cot thu i khong thuoc ind. 
            print('Bien thu %d la bien tu do. ' %(i + 1))
            x[i] = float(input('Ban muon dat bien tu do thu %d bang: ' %(i + 1)))

    # Bat dau qua trinh khu. 
    for i in range(n - 1, -1, -1): 
        if i in ind:   # neu bien thu i + 1 la bien khu (pivot variable).
            for j in range(n - 1, i, -1): 
                if j not in ind: 
                    Ab[row][n] -= Ab[row][j] * x[j]
            
            x[i] = Ab[row][n] / Ab[row][i]
            row -= 1
            
    return x  
```

```python
def GaussJordan(Ab): 
    m = len(Ab)
    n = len(Ab[0]) - 1
    ind = [-1 for i in range(m)]   # ind[i] = j: phan tu khu thu i (hang i) o cot thu j.
    i, j = 0, 0 

    while i <= m - 1 and j <= n - 1:  # dung khi tim duoc het phan tu tru hoac di het cot.
        pivot, pivotRow = pivotSelection(Ab, m, i, j)
        print('pivot%d = %f' % (i + 1, pivot))
        
        if pivot == 0: 
            print('Cột %d không có phần tử khử.' % (j + 1), end='\n\n')
            j += 1
            flag = False 
        else: 
            ind[i] = j  # luu phan tru tru thu i o cot thu j. 
            flag = True 
            
            if pivotRow != i: 
                rowExchange(Ab, n, pivotRow, i)
                print('Row %d and row %d are exchanged!' % (pivotRow + 1, i + 1))
            else: 
                print('Không cần hoán đổi hàng.')
            
        if flag == True: 
            for k in range(m):
                if k != i: 
                    Ab[k][j] = Ab[k][j] / Ab[i][j]   
                
                    for t in range(j + 1, n + 1): 
                        Ab[k][t] = Ab[k][t] - Ab[k][j] * Ab[i][t]

            # dat lai cac he so bang 0 o cot thu j (cho de nhin).
            for t in range(i + 1, m): 
                Ab[t][j] = 0  
            for t in range(i): 
                Ab[t][j] = 0 

            i += 1 
            j += 1  
        
            print('Ma tran Ab sau khi khu phan tu tru thu %d o cot thu %d: ' % (i, j))
            for a in range(m): 
                for b in range(n + 1): 
                    print('%.5f     ' %(Ab[a][b]), end='')
                print()
            print()
          
    printEliMatrix(Ab, m, n, ind, i, j)
    
    rankA, rankAb = rankDeter(Ab, ind, m, n)

    x = Sol(Ab, ind, m, n, rankA, rankAb)
    
    if x == 'He vo nghiem.': 
        print(x)
    else: 
        print('Nghiem cua he Ax = b la: ')
        print(x)
```

## Câu 3: Tính định thức của ma trận vuông dùng Gauss/Gauss Jordan

Câu hỏi: 
- Có cần chọn phần tử khử sao cho giảm sai số đi không? Có.

Chú ý: 
- Nếu trong quá trình khử có trao đổi hàng n lần thì định thức sẽ đổi dấu n lần.
- Nếu có một cột nào đó của ma trận vuông mà không có phần tử khử thì sẽ trả về det(A) = 0. Vì vậy, ta không cần mảng index để theo dõi các cột pivot nữa.
- Viết thêm hàm kiểm tra ma trận vuông nữa.
- In ma trận khử theo các cột thẳng nhau. 

```python
def rowExchange2(A, n, k, i): 
    # Y tuong: di tung cot tu 0 den n, thuc hien trao doi hai phan tu tren hai hang thuoc cot do. 
    for j in range(n): 
        A[k][j], A[i][j] = A[i][j], A[k][j]
```

```python
def printEliMatrix2(A, n):
    print('Elimination augmented matrix: ')
    for i in range(n): 
        for j in range(n): 
            print('%.5f  ' %(A[i][j]), end=' ')
        print()
    print()
```

```python
def pivotSelection2(A, n, i): 
    # Ưu tiên chọn phần tử khử là bội số của 2 theo thứ tự độ lớn tăng dần.
    for k in range(i, n): 
        if abs(A[k][i]) == 1: 
            return A[k][i], k
    for k in range(i, n): 
        if abs(A[k][i]) == 2: 
            return A[k][i], k
    for k in range(i, n): 
        if abs(A[k][i]) == 4: 
            return A[k][i], k
    for k in range(i, n): 
        if A[k][i] != 0 and A[k][i] % 2 == 0: 
            return A[k][i], k
    # Ưu tiên số hai là chọn phần tử khử là phần tử có trị tuyệt đối lớn nhất. 
    max = A[i][i]
    pivotRow = i
    for k in range(i + 1, n): 
        if abs(A[k][i]) > abs(max): 
            max = A[k][i]
            pivotRow = k
    
    return max, pivotRow
```

```python
def GaussDeter(A): 
    n = len(A)
    sign = 1
    
    for i in range(n): 
        pivot, pivotRow = pivotSelection2(A, n, i)
        print('pivot%d = %f' % (i + 1, pivot))
        
        if pivot == 0: 
            print('Cột %d không có phần tử khử.' % (j + 1), end='\n\n')
            flag = False 
        else: 
            flag = True 
            if pivotRow != i: 
                sign *= -1
                rowExchange2(A, n, pivotRow, i)
                print('Row %d and row %d are exchanged!' % (pivotRow + 1, i + 1))
                print('Update sign = %f' % (sign))
            else: 
                print('Không cần hoán đổi hàng.')
            
        if flag == False: 
            return 'detA = 0.0'
        else: 
            for row in range(i + 1, n):   # k là biến hàng
                A[row][i] = A[row][i] / A[i][i] 
                
                for col in range(i + 1, n): # t là biến cột
                    A[row][col] = A[row][col] - A[row][i] * A[i][col]
                
            # Đặt lại các phần tử dưới phần tử đường chéo chính A[i][i] = 0. 
            for row in range(i + 1, n): 
                A[row][i] = 0 

            print('Ma trận A sau khi khử phần tử trụ thứ %d: ' %(i + 1))
            for i in range(n): 
                for j in range(n): 
                    print('%.5f     ' %(A[i][j]), end='')
                print()
            print()
          
    printEliMatrix2(A, n)
    
    # Tính tích của các phần tử trên đường chéo chính của ma trận tam giác trên thu được ở trên. 
    detA = 1
    for i in range(n): 
        detA *= A[i][i]
    detA *= sign 
    
    return 'detA = %f' % (detA)
```

```python
def GaussDetermt(A): 
    n = len(A)
    sign = 1
    
    for i in range(n): 
        flag = True 
        if A[i][i] != 0: 
            print('Không cần hoán đổi.')
        else: 
            print('Cần hoán đổi.')
            
            # Tìm hàng để hoán đổi. 
            for k in range(i + 1, n): 
                if A[k][i] != 0: 
                    rowExchange(A, n, k, i)
                    sign *= -1 
                    
                    print('Hàng %d và hàng %d đã được hoán đổi!' % (k + 1, i + 1))
                    break 
            else: 
                print('Cột thứ %d không có phần tử hoán đổi được.' % (i + 1))
                flag = False
        
        # Kiểm tra xem cột thứ i có phần tử pivot không. 
        if flag == False: 
            return 'detA = 0.0'
        else: 
            for row in range(i + 1, n):   # k là biến hàng
                A[row][i] = A[row][i] / A[i][i] 
                
                for col in range(i + 1, n): # t là biến cột
                    A[row][col] = A[row][col] - A[row][i] * A[i][col]
                
            # Đặt lại các phần tử dưới phần tử đường chéo chính A[i][i] = 0. 
            for row in range(i + 1, n): 
                A[row][i] = 0 

            print('Ma trận A sau khi khử phần tử trụ thứ %d: ' %(i + 1))
            for i in range(n): 
                for j in range(n): 
                    print('%.5f     ' %(A[i][j]), end='')
                print()
            print()
          
    printEliMatrix(A, n)

    # Tính tích của các phần tử trên đường chéo chính của ma trận tam giác trên thu được ở trên. 
    detA = 1
    for i in range(n): 
        detA *= A[i][i]
    detA *= sign 
    
    return 'detA = %f' % (detA)
```

## Câu 4: 

Chú ý: 
- Các phần tử ngoài đường chéo chính được lặp về dần 0, vì vậy vẫn phải cần sai số để dừng vòng lặp.
- Cần chứng minh được sự hội tụ của thuật toán.

```python
def JacobiEigenAlgo(A): # chú ý là ma trận A phải là ma trận đối xứng.
    n = len(A)
    flag = True
    iter = 0
    
    S = np.identity(n)
    
    # Tìm phần tử có trị tuyệt đối lớn nhất ngoài đường chéo chính từ D, ta chỉ search ở nửa trên của D.
    while flag == True:    
        iter += 1
        flag = False
        row, col = 0, 0
        max = 0 
        for i in range(n): 
            for j in range(i + 1, n): 
                if abs(A[i][j]) > max: 
                    max = abs(A[i][j])
                    row, col = i, j
                    
        # Tìm góc quay. 
        if A[row][row] == A[col][col]: 
            if A[row][col] > 0: 
                theta = math.pi / 4
            else: 
                theta = -math.pi / 4
        else: 
            theta = 0.5 * math.atan(2 * A[row][col] / (A[row][row] - A[col][col]))
    
        # Tính ma trận S1 là ma trận quay tương ứng với góc quay theta ở trên. 
        S1 = np.identity(n)
        S1[row][row], S1[col][col] = math.cos(theta), math.cos(theta)
        S1[row][col], S1[col][row] = -math.sin(theta), math.sin(theta)
    
        A = np.matmul(S1.T, np.matmul(A, S1))
        S = np.matmul(S, S1)
        
        # Đặt lại phần tử D[i][j] = 0, vì nếu không sẽ bị xấp xỉ bằng 0. 
        A[row][col], A[col][row] = 0.0, 0.0
        
        # Kiểm tra xem còn phần tử nào lớn hơn 0 không. 
        for i in range(n): 
            for j in range(i + 1, n): 
                if abs(A[i][j]) > 1e-10: 
                    flag = True 
                    break 
            if flag == True: 
                break 
        
        print('iter %d: ' % (iter))
        print('eigenvalues: ')
        for i in range(n): 
            for j in range(n): 
                print('%.10f  ' % (A[i][j]), end=' ')
            print()
        print()
        
        print('eigenvectors: ')
        for i in range(n): 
            for j in range(n): 
                print('%.10f  ' % (S[i][j]), end=' ')
            print()
        print()
        
    return np.diag(A), S
```

Bây giờ, cần sắp xếp lại square_Sigma và V theo thứ tự độ lớn giảm dần. 

```python
def bubbleSort(numbers, arrays): 
    n = len(numbers)
    
    nums = numbers.copy()
    array = arrays.copy()
    
    for i in range(n): 
        for j in range(n - i - 1): 
            if nums[j] < nums[j + 1]: 
                nums[j], nums[j + 1] = nums[j + 1], nums[j]
                array[:, [j, j + 1]] = array[:, [j + 1, j]]
                
    return nums, array
```

### Công thức tìm số k 

```python
def findK(square_Sigma): 
    for k in range(len(square_Sigma)): 
        print('k = %d, r = %f' % (k, np.sum(square_Sigma[:k + 1]) / np.sum(square_Sigma)))
        if np.sum(square_Sigma[:k + 1]) >= 0.98**2 * np.sum(square_Sigma): 
            return k
```

```python
findK(square_Sigma2)
```

```python
def matrixStoring(A): 
    square_Sigma1, V = JacobiEigenAlgo(np.matmul(A.T, A))
    square_Sigma2, U = JacobiEigenAlgo(np.matmul(A, A.T))
    
    square_Sigma1, V = bubbleSort(square_Sigma1, V)
    square_Sigma2, U = bubbleSort(square_Sigma2, U)
    
    print('Sigma1: ')
    print(np.sqrt(square_Sigma1), end='\n\n')
    
    print('V: ')
    print(V, end='\n\n')
    
    print('square_Sigma2: ')
    print(np.sqrt(square_Sigma2), end='\n\n')
    
    print('U: ')
    print(U, end='\n\n')
    
    k = findK(square_Sigma1)
    
    print(k)
    
    return U, np.sqrt(square_Sigma1), V
```
