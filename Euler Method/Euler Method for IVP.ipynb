```python
# Euler an. 

# Euler hien. 

def f(x, y):
    return (x*y)/2

def expEuler(x0, y0, a, b, h):
    n = (b - a)/h   # so khoang (cach deu)
    
    X = []
    Y = []
    X.append(x0)
    Y.append(y0)
    print(x0)
    print(y0)
    print()
    
    xi = x0 
    yi = y0
    for i in np.arange(1, n + 1):
        yi = yi + h*f(xi, yi)
        xi = x0 + i*h
        X.append(xi)
        Y.append(yi)
        
        print(xi)
        print(yi)
        print()
```
