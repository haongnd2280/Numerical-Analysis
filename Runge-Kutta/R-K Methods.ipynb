```python
def f(x, y): 
    return x + y

def main(x0, y0, a, b, h, s): 
    if s == 3: 
        RKthree(x0, y0, a, b, h, s)
    if s == 4: 
        RKfour(x0, y0, a, b, h, s)
```

```python
def RKthree(x0, y0, a, b, h, s): 
    print("Nhap vao p1: ")
    p1 = float(input())
    print("Nhap vao p2: ")
    p2 = float(input())
    print("Nhap vao p3: ")
    p3 = float(input())
    print("Nhap vao alpha2: ")
    alpha2 = float(input())
    print("Nhap vao alpha3: ")
    alpha3 = float(input())
    print("Nhap vao beta21: ")
    beta21 = float(input())
    print("Nhap vao beta31: ")
    beta31 = float(input())
    print("Nhap vao beta32: ")
    beta32 = float(input())
    
    n = (b - a)/h   # so khoang (cach deu).
    x = x0 
    y = y0
    
    for i in range(0, int(n)): 
        k1i = h*f(x, y)
        k2i = h*f(x + alpha2*h, y + beta21*k1i)
        k3i = h*f(x + alpha3*h, y + beta31*k1i + beta32*k2i)
        
        y = y + p1*k1i + p2*k2i + p3*k3i
        x = x0 + (i + 1)*h
        
        print(x)
        print(y)
        
        print()
```
